```python
# Install ipywidgets (required for interactivity in Colab)
!pip install ipywidgets -q
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from scipy.signal import lfilter

# Define a simple high-pass (DC-blocking) filter: first-difference
b = [1, -1]  # y[n] = x[n] - x[n-1]
a = [1]

N = 200
n = np.arange(N)

def plot_signal(mean=0.0):
    np.random.seed(42)  # for reproducible noise
    x = mean + 0.05 * np.random.randn(N)  # input: constant mean + small noise
    y = lfilter(b, a, x)                  # apply high-pass filter

    plt.figure(figsize=(10, 4))
    plt.plot(n, x, label=f'Input (mean = {np.mean(x):.2f})', alpha=0.8)
    plt.plot(n, y, label=f'Output (mean = {np.mean(y):.2f})', linewidth=2)
    plt.axhline(0, color='k', linestyle='--', linewidth=0.7)
    plt.title('High-Pass Filter Removing the Average (DC Component)')
    plt.xlabel('Sample Index')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.ylim(-1.5, 1.5)
    plt.show()

# Create interactive slider
interact(plot_signal,
         mean=FloatSlider(value=0.0, min=-2.0, max=2.0, step=0.1,
                          description='Input Mean:'));
```

```python
# Install only what's needed (ipywidgets for interactivity)
!pip install ipywidgets -q
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Sampling parameters
fs = 100      # Sampling rate (Hz)
T = 2.0       # Duration (seconds)
N = int(T * fs)
t = np.linspace(0, T, N, endpoint=False)

def remove_dc_and_plot(mean=1.0):
    # Create input signal: DC + sine waves + noise
    np.random.seed(42)
    x = (mean +
         0.5 * np.sin(2 * np.pi * 2 * t) +    # 2 Hz tone
         0.3 * np.sin(2 * np.pi * 20 * t) +   # 20 Hz tone
         0.1 * np.random.randn(N))            # small noise

    # Simple high-pass (DC-removal): first difference
    y = np.zeros_like(x)
    y[1:] = x[1:] - x[:-1]  # y[n] = x[n] - x[n-1]

    # Compute FFTs (magnitude spectrum)
    X_fft = np.fft.rfft(x)
    Y_fft = np.fft.rfft(y)
    freqs = np.fft.rfftfreq(N, 1/fs)

    # Plot
    fig, axs = plt.subplots(2, 2, figsize=(14, 8))

    # Time domain
    axs[0, 0].plot(t, x, 'b')
    axs[0, 0].set_title(f'Input Signal (Mean = {np.mean(x):.2f})')
    axs[0, 0].set_xlabel('Time (s)')
    axs[0, 0].grid(True)

    axs[0, 1].plot(t, y, 'r')
    axs[0, 1].set_title(f'Output Signal (Mean = {np.mean(y):.2f})')
    axs[0, 1].set_xlabel('Time (s)')
    axs[0, 1].grid(True)

    # Frequency domain (log scale for clarity)
    axs[1, 0].semilogy(freqs, np.abs(X_fft), 'b')
    axs[1, 0].set_title('Input Spectrum')
    axs[1, 0].set_xlabel('Frequency (Hz)')
    axs[1, 0].set_xlim(0, 30)
    axs[1, 0].grid(True)

    axs[1, 1].semilogy(freqs, np.abs(Y_fft), 'r')
    axs[1, 1].set_title('Output Spectrum')
    axs[1, 1].set_xlabel('Frequency (Hz)')
    axs[1, 1].set_xlim(0, 30)
    axs[1, 1].grid(True)

    plt.tight_layout()
    plt.show()

# Interactive slider
interact(remove_dc_and_plot,
         mean=FloatSlider(value=1.0, min=-3.0, max=3.0, step=0.1,
                          description='Input Mean:'));
```

```python
# Install only what's needed (ipywidgets for interactivity)
!pip install ipywidgets -q
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Sampling parameters
fs = 100      # Sampling rate (Hz)
T = 2.0       # Duration (seconds)
N = int(T * fs)
t = np.linspace(0, T, N, endpoint=False)

def remove_dc_and_plot(mean=1.0):
    # Create input signal: DC + sine waves + noise
    np.random.seed(42)
    x = (mean +
         0.5 * np.sin(2 * np.pi * 2 * t) +    # 2 Hz tone
         0.3 * np.sin(2 * np.pi * 20 * t) +   # 20 Hz tone
         0.1 * np.random.randn(N))            # small noise

    # Simple high-pass (DC-removal): first difference
    y = np.zeros_like(x)
    y[1:] = x[1:] - x[:-1]  # y[n] = x[n] - x[n-1]

    # Compute FFTs (magnitude spectrum)
    X_fft = np.fft.rfft(x)
    Y_fft = np.fft.rfft(y)
    freqs = np.fft.rfftfreq(N, 1/fs)

    # Compute autocorrelations
    Rx = np.correlate(x, x, mode='full')
    Ry = np.correlate(y, y, mode='full')
    lags = np.arange(-N + 1, N)

    # Plot: 3 rows (time, spectrum, autocorr), 2 columns (input, output)
    fig, axs = plt.subplots(3, 2, figsize=(14, 10))

    # Time domain
    axs[0, 0].plot(t, x, 'b')
    axs[0, 0].set_title(f'Input Signal (Mean = {np.mean(x):.2f})')
    axs[0, 0].set_xlabel('Time (s)')
    axs[0, 0].grid(True)

    axs[0, 1].plot(t, y, 'r')
    axs[0, 1].set_title(f'Output Signal (Mean = {np.mean(y):.2f})')
    axs[0, 1].set_xlabel('Time (s)')
    axs[0, 1].grid(True)

    # Frequency domain (log scale for clarity)
    axs[1, 0].semilogy(freqs, np.abs(X_fft), 'b')
    axs[1, 0].set_title('Input Spectrum')
    axs[1, 0].set_xlabel('Frequency (Hz)')
    axs[1, 0].set_xlim(0, 30)
    axs[1, 0].grid(True)

    axs[1, 1].semilogy(freqs, np.abs(Y_fft), 'r')
    axs[1, 1].set_title('Output Spectrum')
    axs[1, 1].set_xlabel('Frequency (Hz)')
    axs[1, 1].set_xlim(0, 30)
    axs[1, 1].grid(True)

    # Autocorrelation (Rx and Ry)
    axs[2, 0].plot(lags, Rx, 'b')
    axs[2, 0].set_title('Input Autocorrelation $R_x(\\tau)$')
    axs[2, 0].set_xlabel('Lag $\\tau$ (samples)')
    axs[2, 0].grid(True)

    axs[2, 1].plot(lags, Ry, 'r')
    axs[2, 1].set_title('Output Autocorrelation $R_y(\\tau)$')
    axs[2, 1].set_xlabel('Lag $\\tau$ (samples)')
    axs[2, 1].grid(True)

    plt.tight_layout()
    plt.show()

# Interactive slider
interact(remove_dc_and_plot,
         mean=FloatSlider(value=1.0, min=-3.0, max=3.0, step=0.1,
                          description='Input Mean:'));
```

```python
# Instalar ipywidgets si es necesario
!pip install ipywidgets -q
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Parmetros del sistema
R = 1e3      # 1 k
C = 1e-6     # 1 F
a = 1 / (R * C)  # a = 1/RC

# Simulacin
fs = 1000    # Hz
T = 5.0      # segundos
N = int(T * fs)
t = np.linspace(0, T, N, endpoint=False)

def simular_no_estacionario(alpha=1.0):
    """
    alpha: escala para m1(t)  en la imagen, alpha=1 y alpha=2
    """
    # --- 1. Definir media de entrada m1(t) ---
    m1xi = np.where(t < 1.5, 1.0, alpha)  # Cambio en t=1.5s

    # --- 2. Generar seal de entrada (t) como ruido con media variable ---
    np.random.seed(42)
    noise = 0.1 * np.random.randn(N)  # Ruido pequeo
    xi = m1xi + noise  # Entrada con media variable

    # --- 3. Calcular salida (t) mediante la integral ---
    eta = np.zeros_like(t)
    dt = t[1] - t[0]

    for i in range(1, N):
        # Integral desde 0 hasta t_i:  e^(a) () d
        integral = np.trapz(np.exp(a * t[:i]) * xi[:i], t[:i])
        eta[i] = a * np.exp(-a * t[i]) * integral

    # --- 4. Calcular media terica E{(t)} ---
    # Segn la frmula: E{(t)} = m1(t) * a * e^(-at) * (e^(at) - 1)
    # Pero m1(t) es variable  usamos la expresin integral
    mean_eta_theoretical = np.zeros_like(t)
    for i in range(1, N):
        integral_mean = np.trapz(np.exp(a * t[:i]) * m1xi[:i], t[:i])
        mean_eta_theoretical[i] = a * np.exp(-a * t[i]) * integral_mean

    # --- Grficos ---
    fig, axs = plt.subplots(3, 1, figsize=(12, 9))

    # Entrada (t)
    axs[0].plot(t, xi, 'b-', linewidth=1, label='Entrada (t)')
    axs[0].plot(t, m1xi, 'r--', linewidth=2, label='Media de entrada m(t)')
    axs[0].set_title('Seal de Entrada (t) con Media Variable')
    axs[0].set_xlabel('Tiempo (s)')
    axs[0].set_ylabel('Amplitud')
    axs[0].grid(True, linestyle=':', alpha=0.7)
    axs[0].legend()

    # Salida (t)
    axs[1].plot(t, eta, 'g-', linewidth=2, label='Salida (t)')
    axs[1].plot(t, mean_eta_theoretical, 'm--', linewidth=2, label='Media terica E{(t)}')
    axs[1].set_title('Salida (t) y su Media Terica')
    axs[1].set_xlabel('Tiempo (s)')
    axs[1].set_ylabel('Amplitud')
    axs[1].grid(True, linestyle=':', alpha=0.7)
    axs[1].legend()

    # Zoom en la regin donde cambia la media
    axs[2].plot(t, eta, 'g-', linewidth=2, label='Salida (t)')
    axs[2].plot(t, mean_eta_theoretical, 'm--', linewidth=2, label='Media terica E{(t)}')
    axs[2].axvline(x=1.5, color='k', linestyle='--', linewidth=1, label='Cambio en m(t)')
    axs[2].set_xlim(0.5, 3.5)
    axs[2].set_title('Zoom: Respuesta al Cambio de Media (t=1.5s)')
    axs[2].set_xlabel('Tiempo (s)')
    axs[2].set_ylabel('Amplitud')
    axs[2].grid(True, linestyle=':', alpha=0.7)
    axs[2].legend()

    plt.tight_layout()
    plt.show()

# Slider interactivo para cambiar alpha (como en la imagen: alpha=1 o 2)
interact(simular_no_estacionario,
         alpha=FloatSlider(value=1.0, min=1.0, max=3.0, step=0.5,
                           description='alpha (m despus de 1.5s):'));
```
