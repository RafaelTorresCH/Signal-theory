# **Pulse Modulation Schemes in Signal Theory**

## Pulse Amplitude Modulation (PAM)

In PAM, the **amplitude** of each pulse is proportional to the instantaneous value of the modulating signal $x(t)$.

$$
s_{\text{PAM}}(t) = \sum_{n=-\infty}^{\infty} x(nT_s) \, p(t - nT_s)
$$

Where:  
- $x(t)$ â†’ analog modulating signal  
- $p(t)$ â†’ unit pulse of duration $\tau_p$  
- $T_s$ â†’ sampling period  

---

##  Pulse Width Modulation (PWM)

In PWM, the **width** (duration) of each pulse is varied in proportion to the amplitude of $x(t)$.

$$
s_{\text{PWM}}(t) = \sum_{n=-\infty}^{\infty} A_p \, p\!\big(t - nT_s; \, \tau_n\big)
$$

with pulse width  

$$
\tau_n = \tau_0 + k_w \, x(nT_s)
$$

Where:  
- $A_p$ â†’ constant pulse amplitude  
- $\tau_n$ â†’ instantaneous pulse width  
- $k_w$ â†’ sensitivity factor (sec/unit amplitude)  

---

## Pulse Position Modulation (PPM)

In PPM, the **time position** of each pulse is shifted relative to a reference clock according to $x(t)$.

$$
s_{\text{PPM}}(t) = \sum_{n=-\infty}^{\infty} A_p \, p\!\big(t - nT_s - k_p \, x(nT_s)\big)
$$

Where:  
- $A_p$ â†’ constant pulse amplitude  
- $k_p$ â†’ position sensitivity (sec/unit amplitude)  
- The pulse position within each period $T_s$ carries the information of $x(t)$  

---

##  Pulse Code Modulation (PCM)

In PCM, each sample $x(t)$ is **quantized and encoded** into a digital word of $N$ bits.

1. **Sampling:**  
   $$
   x_s(t) = \sum_{n=-\infty}^{\infty} x(nT_s) \, \delta(t - nT_s)
   $$

2. **Quantization:**  
   $$
   x_q(nT_s) = Q\{x(nT_s)\}
   $$

3. **Encoding:**  
   Each quantized level is represented by a binary sequence  
   $$
   b_n = \text{encode}\big(x_q(nT_s)\big)
   $$

Hence, the PCM signal can be represented abstractly as  

$$
s_{\text{PCM}}(t) = \sum_{n=-\infty}^{\infty} b_n \, p(t - nT_b)
$$

Where:  
- $Q\{\cdot\}$ â†’ quantization operator  
- $b_n$ â†’ binary-coded amplitude (digital word)  
- $T_b$ â†’ bit duration  

---

 **Summary Table**

| Modulation | Varies | Mathematical Relation | Nature |
|:------------|:--------|:----------------------|:--------|
| PAM | Amplitude | $A_n = x(nT_s)$ | Analog |
| PWM | Width | $\tau_n = \tau_0 + k_w x(nT_s)$ | Analog |
| PPM | Position | $t_n = nT_s + k_p x(nT_s)$ | Analog |
| PCM | Code | $b_n = \text{encode}(Q\{x(nT_s)\})$ | Digital |


# **Spectral Expressions for PAM, PWM, PPM, and PCM**

Let the message be $x(t)$, the sampling period $T_s$, and a unit-amplitude pulse shape $p(t)$ with Fourier transform $P(f)=\mathcal{F}\{p(t)\}$.
We use the impulse train $i_{T_s}(t)=\sum_{n\in\mathbb{Z}}\delta(t-nT_s)$, whose spectrum is
$$
\mathcal{F}\{i_{T_s}(t)\}=\frac{1}{T_s}\sum_{k\in\mathbb{Z}}\delta\!\left(f-\frac{k}{T_s}\right).
$$

---

## PAM â€” Pulse Amplitude Modulation

**Signal model**
$$
s_{\text{PAM}}(t)=\sum_{n\in\mathbb{Z}} x(nT_s)\,p\!\big(t-nT_s\big)
\;\;=\;\big[x(t)\,i_{T_s}(t)\big]*p(t).
$$

**Spectrum**
$$
S_{\text{PAM}}(f)=\frac{1}{T_s}\sum_{k\in\mathbb{Z}} X\!\left(f-\frac{k}{T_s}\right)\,P(f).
$$

**(If $x(t)$ is WSS with PSD $\Phi_x(f)$)**
$$
\Phi_{s,\text{PAM}}(f)=\frac{1}{T_s}\sum_{k\in\mathbb{Z}} \Phi_x\!\left(f-\frac{k}{T_s}\right)\,|P(f)|^2.
$$

---

## PWM â€” Pulse Width Modulation

Let pulses have constant amplitude $A_p$, but **width** $\tau_n=\tau_0+k_w\,x(nT_s)$ within the $n$-th slot.
A single rectangular pulse of width $\tau$ centered at $t=0$ has spectrum
$$
\mathcal{F}\{A_p\,\text{rect}(t/\tau)\}
= A_p\,\tau\,\mathrm{sinc}(\pi f\tau)\,e^{-j\pi f\tau}.
$$

**Exact line-by-line superposition**
$$
S_{\text{PWM}}(f)
=\sum_{n\in\mathbb{Z}} e^{-j2\pi f nT_s}\;A_p\,\tau_n\,
\mathrm{sinc}\!\big(\pi f\tau_n\big)\,e^{-j\pi f\tau_n}.
$$

**Small-signal (first-order) linearization** around $\tau_0$:
write $\tau_n=\tau_0+\Delta\tau_n$ with $\Delta\tau_n=k_w x(nT_s)$ and Taylor-expand the pulse spectrum:
$$
S_{\text{PWM}}(f)\;\approx\;
\underbrace{A_p\,\tau_0\,\mathrm{sinc}(\pi f\tau_0)\,e^{-j\pi f\tau_0}}_{\displaystyle C_0(f)}
\sum_{n}e^{-j2\pi f nT_s}
\;+\;
\underbrace{A_p\,\frac{\partial}{\partial \tau}\!\Big[\tau\,\mathrm{sinc}(\pi f\tau)\,e^{-j\pi f\tau}\Big]_{\tau=\tau_0}}_{\displaystyle G_{\text{PWM}}(f)}
\sum_{n}\Delta\tau_n\,e^{-j2\pi f nT_s}.
$$

Using Poisson summation on the data term
$\sum_n \Delta\tau_n e^{-j2\pi f nT_s}
= \frac{k_w}{T_s}\sum_{k} X\!\left(f-\frac{k}{T_s}\right)$,
the **modulation sidebands** ride on replicas of $X(f)$ scaled by $G_{\text{PWM}}(f)$.

---

##  PPM â€” Pulse Position Modulation

Pulses have **fixed shape** $p(t)$ and **fixed amplitude** $A_p$, but are **time-shifted** by
$\Delta t_n=k_p\,x(nT_s)$ within each slot:
$$
s_{\text{PPM}}(t)=\sum_{n\in\mathbb{Z}} A_p\,p\!\big(t-nT_s-\Delta t_n\big).
$$

**Exact spectrum (nonlinear phase term)**
$$
S_{\text{PPM}}(f)
= A_p\,P(f)\sum_{n\in\mathbb{Z}} e^{-j2\pi f (nT_s+\Delta t_n)}
= A_p\,P(f)\sum_{n} e^{-j2\pi f nT_s}\,e^{-j2\pi f k_p x(nT_s)}.
$$

**Small-signal (first-order) linearization** with $|2\pi f k_p x(nT_s)|\ll 1$:
$$
e^{-j2\pi f k_p x(nT_s)}\approx 1 - j2\pi f k_p\,x(nT_s),
$$
so
$$
S_{\text{PPM}}(f)\;\approx\;A_p\,P(f)\sum_{n} e^{-j2\pi f nT_s}
\;-\;j\,2\pi f\,k_p\,A_p\,P(f)\sum_{n} x(nT_s)\,e^{-j2\pi f nT_s}.
$$

By Poisson summation,
$$
\sum_{n} x(nT_s)\,e^{-j2\pi f nT_s}
=\frac{1}{T_s}\sum_{k\in\mathbb{Z}} X\!\left(f-\frac{k}{T_s}\right),
$$
hence **PPM â‰ˆ phase modulation** of the pulse train, producing sidebands proportional to $f\,P(f)$ and the spectrum of $x(t)$.

---

##  PCM â€” Pulse Code Modulation (line-coded)

Let a **bit (or symbol) sequence** $\{b_n\}$ drive a pulse shape $p_b(t)$ every $T_b$:
$$
s_{\text{PCM}}(t)=\sum_{n\in\mathbb{Z}} b_n\,p_b\!\big(t-nT_b\big),\qquad
P_b(f)=\mathcal{F}\{p_b(t)\}.
$$

**Deterministic spectrum (DTFT view)**
$$
S_{\text{PCM}}(f)=P_b(f)\;\underbrace{\sum_{n} b_n\,e^{-j2\pi f nT_b}}_{\displaystyle B\!\big(e^{-j2\pi f T_b}\big)}.
$$

**Power Spectral Density (wide-sense stationary $\{b_n\}$)**
Let $\Psi_b(e^{j\omega})$ be the discrete-time PSD of $\{b_n\}$. Then the continuous-time PSD is
$$
\Phi_{s,\text{PCM}}(f)\;=\;\frac{|P_b(f)|^2}{T_b}\;\Psi_b\!\big(e^{j2\pi f T_b}\big).
$$

**Common NRZ rectangular pulse (amplitude $A$, width $T_b$)**
$$
p_b(t)=A\,\text{rect}\!\left(\frac{t}{T_b}\right)
\;\Rightarrow\;
P_b(f)=A\,T_b\,\mathrm{sinc}(\pi f T_b)\,e^{-j\pi f T_b}.
$$

**IID binary case** (mean $m=\mathbb{E}[b_n]$, variance $\sigma_b^2$):
$$
\Psi_b\!\big(e^{j\omega}\big)=\sigma_b^2 \;+\; 2\pi\,m^2\,\sum_{k\in\mathbb{Z}}\delta(\omega-2\pi k),
$$
thus
$$
\Phi_{s,\text{PCM}}(f)=\frac{|P_b(f)|^2}{T_b}\,\sigma_b^2
\;+\;|P_b(f)|^2\,m^2\sum_{k\in\mathbb{Z}}\delta\!\left(f-\frac{k}{T_b}\right).
$$

> *Interpretation:* the continuous part comes from randomness ($\sigma_b^2$), while discrete lines at multiples of $1/T_b$ appear if the sequence has nonzero mean ($m\neq 0$). Line codes with DC suppression (e.g., Manchester) remove the $m^2$ line at $f=0$.

---


# **Digital Spectra of PAM, PWM, PPM, and PCM**

Let:
- $x[n]$ = discrete-time (sampled) message
- Symbols occur every $M$ samples (symbol period $M$)
- $p[n]$ = discrete-time pulse shape with DTFT $P(e^{j\omega})$
- The **upsampled** sequence by $M$ is $u[n]=\sum_{m\in\mathbb{Z}} x[m]\;\delta[n-mM]$
  with DTFT $\mathcal{F}\{u[n]\}=\frac{1}{M}\sum_{k\in\mathbb{Z}} X\!\left(e^{j(\omega-2\pi k/M)}\right)$

---

## ðŸ”¹ Digital PAM (discrete-time)

**Signal model**
$$
s_{\text{PAM}}[n] \;=\; (u * p)[n] \;=\; \sum_{m\in\mathbb{Z}} x[m]\;p[n-mM].
$$

**Spectrum (DTFT)**
$$
S_{\text{PAM}}\!\left(e^{j\omega}\right)
=\frac{1}{M}\sum_{k\in\mathbb{Z}} X\!\left(e^{j(\omega-2\pi k/M)}\right)\;P\!\left(e^{j\omega}\right).
$$

**If $x[n]$ is WSS with discrete-time PSD $\Psi_x(e^{j\omega})$**
$$
\Psi_{s,\text{PAM}}\!\left(e^{j\omega}\right)
=\frac{1}{M}\sum_{k\in\mathbb{Z}} \Psi_x\!\left(e^{j(\omega-2\pi k/M)}\right)\;\big|P(e^{j\omega})\big|^2.
$$

---

## ðŸ”¹ Digital PWM (discrete-time, slot length $M$)

Within the $m$-th symbol slot, use a **rectangular pulse** of **integer width** $L_m$ samples:
$$
L_m \;=\; L_0 \;+\; k_w\,x[m],
\quad s_{\text{PWM}}[n]\;=\;\sum_{m\in\mathbb{Z}} A_p\,\text{rect}\!\big(n-mM;L_m\big),
$$
where $\text{rect}(n;L)=1$ for $n\in\{0,1,\dots,L-1\}$ and $0$ otherwise.
The DTFT of a length-$L$ rectangle is the **Dirichlet kernel**
$$
\mathcal{F}\{\text{rect}(n;L)\}
\;=\; e^{-j\omega(L-1)/2}\;\frac{\sin\!\big(\tfrac{L\,\omega}{2}\big)}{\sin\!\big(\tfrac{\omega}{2}\big)}.
$$

**Exact line-by-line superposition**
$$
S_{\text{PWM}}\!\left(e^{j\omega}\right)
=\sum_{m\in\mathbb{Z}} e^{-j\omega mM}\;
A_p\;e^{-j\omega(L_m-1)/2}\;
\frac{\sin\!\big(\tfrac{L_m\,\omega}{2}\big)}{\sin\!\big(\tfrac{\omega}{2}\big)}.
$$

**Small-signal linearization** ($L_m=L_0+\Delta L_m$, $\Delta L_m=k_w x[m]$, first-order Taylor):
Define
$$
D(\omega,L)=e^{-j\omega(L-1)/2}\;\frac{\sin\!\big(\tfrac{L\,\omega}{2}\big)}{\sin\!\big(\tfrac{\omega}{2}\big)},\qquad
G_{\text{PWM}}(\omega)=A_p\,\left.\frac{\partial D(\omega,L)}{\partial L}\right|_{L=L_0}.
$$
Then
$$
S_{\text{PWM}}\!\left(e^{j\omega}\right)\;\approx\;
\underbrace{A_p\,D(\omega,L_0)\sum_{m}e^{-j\omega mM}}_{\text{clock lines at } \omega=2\pi k/M}
\;+\; G_{\text{PWM}}(\omega)\sum_{m}\Delta L_m\,e^{-j\omega mM}.
$$
By upsampling identity,
$$
\sum_{m}\Delta L_m\,e^{-j\omega mM}
=\frac{k_w}{M}\sum_{k\in\mathbb{Z}}X\!\left(e^{j(\omega-2\pi k/M)}\right),
$$
i.e., **PWM sidebands** are aliases of $X(e^{j\omega})$ shaped by $G_{\text{PWM}}(\omega)$ around the clock harmonics.

---

## ðŸ”¹ Digital PPM (discrete-time, slot length $M$)

Fixed pulse shape $p[n]$, **time shift (in samples)** inside each slot:
$$
\Delta_m\;=\;k_p\,x[m],\qquad
s_{\text{PPM}}[n]=\sum_{m\in\mathbb{Z}} A_p\,p[n-mM-\Delta_m].
$$

**Spectrum (exact)**
$$
S_{\text{PPM}}\!\left(e^{j\omega}\right)=A_p\,P\!\left(e^{j\omega}\right)\sum_{m\in\mathbb{Z}} e^{-j\omega mM}\,e^{-j\omega \Delta_m}.
$$

**Small-signal linearization** ($|\omega\,\Delta_m|\ll 1$):
$$
e^{-j\omega\Delta_m}\approx 1 - j\omega\,\Delta_m
\quad\Rightarrow\quad
S_{\text{PPM}}\!\left(e^{j\omega}\right)\approx
A_p\,P\!\left(e^{j\omega}\right)\sum_{m} e^{-j\omega mM}
\;-\; j\omega k_p A_p\,P\!\left(e^{j\omega}\right)\sum_{m} x[m]\,e^{-j\omega mM}.
$$
Using the upsampling identity:
$$
\sum_{m} x[m]\,e^{-j\omega mM}
=\frac{1}{M}\sum_{k\in\mathbb{Z}} X\!\left(e^{j(\omega-2\pi k/M)}\right).
$$
Thus, **PPM â‰ˆ phase modulation** of the pulse train: sidebands $\propto\;\omega\,P(e^{j\omega})$ times aliased $X(e^{j\omega})$.

---

## ðŸ”¹ Digital PCM (line-coded, discrete-time)

Let a **symbol/bit sequence** $\{b[m]\}$ (WSS) be upsampled by $M$ and shaped by $p_b[n]$:
$$
s_{\text{PCM}}[n]=\sum_{m\in\mathbb{Z}} b[m]\;p_b[n-mM],
\qquad P_b(e^{j\omega})=\mathcal{F}\{p_b[n]\}.
$$

**Spectrum**
$$
S_{\text{PCM}}\!\left(e^{j\omega}\right)
= P_b\!\left(e^{j\omega}\right)\;\frac{1}{M}\sum_{k\in\mathbb{Z}} B\!\left(e^{j(\omega-2\pi k/M)}\right),
$$
where $B(e^{j\omega})$ is the DTFT of $\{b[m]\}$.

**PSD (if $\{b[m]\}$ is WSS with discrete-time PSD $\Psi_b(e^{j\omega})$)**
$$
\Psi_{s,\text{PCM}}\!\left(e^{j\omega}\right)
=\frac{|P_b(e^{j\omega})|^2}{M}\;\sum_{k\in\mathbb{Z}}
\Psi_b\!\left(e^{j(\omega-2\pi k/M)}\right).
$$

**Example: NRZ rectangular shaping** of length $M$ at amplitude $A$:
$$
p_b[n]=A\,\text{rect}(n;M)
\;\Rightarrow\;
P_b(e^{j\omega})=A\,e^{-j\omega(M-1)/2}\,\frac{\sin(M\omega/2)}{\sin(\omega/2)}.
$$
For i.i.d. binary with mean $m$ and variance $\sigma_b^2$:
$$
\Psi_b(e^{j\omega})=\sigma_b^2 \;+\;2\pi\,m^2\sum_{\ell\in\mathbb{Z}}\delta(\omega-2\pi\ell),
$$
so DC (and symbol-rate) **lines** appear when $m\neq 0$, while the **continuous** part scales with $\sigma_b^2$ and $|P_b|^2$.

---


```python
# One-figure spectra for PAM, PWM, PPM, and PCM (digital, discrete-time)
# Uses numpy & matplotlib only. Creates a single PNG with all four spectra.

import numpy as np
import matplotlib.pyplot as plt

# ----------------------
# Utility helpers
# ----------------------
def next_pow2(n):
    return 1 << (n - 1).bit_length()

def fft_spectrum(x, zero_pad=True):
    N = len(x)
    Nfft = next_pow2(N) if zero_pad else N
    X = np.fft.fftshift(np.fft.fft(x, Nfft))
    w = np.linspace(-np.pi, np.pi, Nfft, endpoint=False)  # rad/sample
    mag = np.abs(X) + 1e-15
    mag_db = 20*np.log10(mag / np.max(mag))
    return w, mag_db

def rect_pulse(L):
    p = np.zeros(L, dtype=float)
    p[:] = 1.0
    return p

# ----------------------
# Message & framing params
# ----------------------
np.random.seed(7)
Nsym = 64        # number of symbols
M    = 16        # samples per symbol (slot length)

# Baseband message x[m]: lowpass-like (sum of sines) for clarity
m = np.arange(Nsym)
x_msg = 0.6*np.sin(2*np.pi*0.05*m) + 0.3*np.sin(2*np.pi*0.12*m)

# Common pulse shape (discrete-time) for PAM/PPM (rectangular)
Lp = 8                           # pulse length (<= M preferred)
p = rect_pulse(Lp)

# ----------------------
# 1) Digital PAM
# s[n] = sum_m x[m] * p[n - mM]
# ----------------------
pam = np.zeros(Nsym*M + Lp, dtype=float)
for i, xm in enumerate(x_msg):
    start = i*M
    pam[start:start+Lp] += xm * p

w_pam, S_pam = fft_spectrum(pam)

# ----------------------
# 2) Digital PWM
# Width L_m = L0 + k_w * x[m], clamped to [1, M]
# s[n] has a length-L_m rectangle at the beginning of each slot
# ----------------------
L0   = max(2, Lp)   # nominal width (samples)
k_w  = int(0.25*M)  # sensitivity in samples per unit amplitude

pwm = np.zeros(Nsym*M, dtype=float)
for i, xm in enumerate(x_msg):
    Lm = int(np.round(L0 + k_w * xm))
    Lm = max(1, min(M, Lm))
    start = i*M
    pwm[start:start+Lm] += 1.0  # amplitude Ap = 1

w_pwm, S_pwm = fft_spectrum(pwm)

# ----------------------
# 3) Digital PPM
# Time shift Î”_m = k_p * x[m] (in samples, integer), inside slot
# s[n] places the fixed pulse p[n] shifted by Î”_m within each slot
# ----------------------
k_p = int(0.25*Lp)  # shift sensitivity (samples per unit amplitude)

ppm = np.zeros(Nsym*M + Lp, dtype=float)
for i, xm in enumerate(x_msg):
    shift = int(np.round(k_p * xm))
    # center pulse within slot, then shift by Î”_m and clamp to slot
    base = i*M + (M - Lp)//2 + shift
    base = max(i*M, min(i*M + M - Lp, base))
    ppm[base:base+Lp] += p  # amplitude Ap = 1

w_ppm, S_ppm = fft_spectrum(ppm)

# ----------------------
# 4) Digital PCM (line-coded NRZ)
# Quantize x[m] to Q levels, map to bipolar symbols b[m], then shape with NRZ pulse of length M
# ----------------------
Qbits = 4
Qlev  = 2**Qbits
xmin, xmax = -1.0, 1.0
xq = np.clip(x_msg, xmin, xmax)
xq = np.round((xq - xmin)/(xmax - xmin)*(Qlev - 1)).astype(int)  # 0..Qlev-1

# Map to symmetric levels in [-1, +1]
b_sym = (xq/(Qlev - 1))*2 - 1
# Optional: remove mean to suppress DC line
# b_sym = b_sym - np.mean(b_sym)

p_b = rect_pulse(M)  # NRZ pulse
pcm = np.zeros(Nsym*M + M, dtype=float)
for i, bm in enumerate(b_sym):
    start = i*M
    pcm[start:start+M] += bm * p_b

w_pcm, S_pcm = fft_spectrum(pcm)

# ----------------------
# Single FIGURE with 4 subplots
# ----------------------
fig, axes = plt.subplots(4, 1, figsize=(10, 14), sharex=True)
axes[0].plot(w_pam, S_pam)
axes[0].set_title("Digital PAM â€” |S(e^{jÏ‰})| (dB)")
axes[0].set_ylabel("Magnitude (dB)")
axes[0].grid(True)

axes[1].plot(w_pwm, S_pwm)
axes[1].set_title("Digital PWM â€” |S(e^{jÏ‰})| (dB)")
axes[1].set_ylabel("Magnitude (dB)")
axes[1].grid(True)

axes[2].plot(w_ppm, S_ppm)
axes[2].set_title("Digital PPM â€” |S(e^{jÏ‰})| (dB)")
axes[2].set_ylabel("Magnitude (dB)")
axes[2].grid(True)

axes[3].plot(w_pcm, S_pcm)
axes[3].set_title("Digital PCM (NRZ shaping) â€” |S(e^{jÏ‰})| (dB)")
axes[3].set_xlabel("Ï‰ (rad/sample)")
axes[3].set_ylabel("Magnitude (dB)")
axes[3].grid(True)

fig.tight_layout()
out_path = "/content/digital_spectra_all.png"
plt.savefig(out_path, dpi=150, bbox_inches="tight")
plt.show()

print("Saved:", out_path)
```

```python
# ============================
# CÃ“DIGO A â€” PSD con Welch
# ============================
# Digital PAM, PWM, PPM, PCM + Welch PSD
# Requisitos: numpy, matplotlib, scipy

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# ---------- utilidades ----------
def next_pow2(n):
    return 1 << (n - 1).bit_length()

def rect_pulse(L):
    return np.ones(L, dtype=float)

def fftfreq_rad(N):
    # Frecuencia digital en rad/muestra para graficar PSD en Ï‰ si lo deseas
    return np.linspace(0, np.pi, N//2+1)

# ---------- parÃ¡metros base ----------
np.random.seed(7)
Nsym = 128         # sÃ­mbolos
M    = 16          # muestras por sÃ­mbolo
Lp   = 8           # longitud del pulso rectangular (<= M)
fs   = 1.0         # Hz (discreto); para PSD de Welch no afecta forma si es constante

# mensaje discreto (suave) para ver bandas con claridad
m = np.arange(Nsym)
x_msg = 0.6*np.sin(2*np.pi*0.045*m) + 0.35*np.sin(2*np.pi*0.12*m)

# ---------- 1) PAM ----------
p = rect_pulse(Lp)
pam = np.zeros(Nsym*M + Lp, dtype=float)
for i, xm in enumerate(x_msg):
    start = i*M
    pam[start:start+Lp] += xm * p

f_pam, Pxx_pam = welch(pam, fs=fs, nperseg=256, noverlap=128, nfft=1024, return_onesided=True, scaling='density')

# ---------- 2) PWM ----------
L0   = max(2, Lp)        # ancho nominal
k_w  = int(0.25*M)       # sensibilidad (muestras por unidad)
pwm = np.zeros(Nsym*M, dtype=float)
for i, xm in enumerate(x_msg):
    Lm = int(np.round(L0 + k_w * xm))
    Lm = max(1, min(M, Lm))
    start = i*M
    pwm[start:start+Lm] += 1.0

f_pwm, Pxx_pwm = welch(pwm, fs=fs, nperseg=256, noverlap=128, nfft=1024, return_onesided=True, scaling='density')

# ---------- 3) PPM ----------
k_p = int(0.25*Lp)   # desplazamiento (muestras por unidad)
ppm = np.zeros(Nsym*M + Lp, dtype=float)
for i, xm in enumerate(x_msg):
    shift = int(np.round(k_p * xm))
    base = i*M + (M - Lp)//2 + shift
    base = max(i*M, min(i*M + M - Lp, base))
    ppm[base:base+Lp] += p

f_ppm, Pxx_ppm = welch(ppm, fs=fs, nperseg=256, noverlap=128, nfft=1024, return_onesided=True, scaling='density')

# ---------- 4) PCM (NRZ line coding) ----------
Qbits = 4
Qlev  = 2**Qbits
xmin, xmax = -1.0, 1.0
xq = np.clip(x_msg, xmin, xmax)
xq = np.round((xq - xmin)/(xmax - xmin)*(Qlev - 1)).astype(int)    # 0..Qlev-1
b_sym = (xq/(Qlev - 1))*2 - 1                                      # map a [-1,1]
# opcional: sacar media para suprimir DC
# b_sym = b_sym - np.mean(b_sym)

p_b = rect_pulse(M)  # NRZ
pcm = np.zeros(Nsym*M + M, dtype=float)
for i, bm in enumerate(b_sym):
    start = i*M
    pcm[start:start+M] += bm * p_b

f_pcm, Pxx_pcm = welch(pcm, fs=fs, nperseg=256, noverlap=128, nfft=1024, return_onesided=True, scaling='density')

# ---------- PLOT EN UNA SOLA FIGURA ----------
fig, axes = plt.subplots(4, 1, figsize=(10, 14), sharex=True)
axes[0].semilogy(f_pam, Pxx_pam); axes[0].grid(True); axes[0].set_title("PAM â€” Welch PSD"); axes[0].set_ylabel("PSD")
axes[1].semilogy(f_pwm, Pxx_pwm); axes[1].grid(True); axes[1].set_title("PWM â€” Welch PSD"); axes[1].set_ylabel("PSD")
axes[2].semilogy(f_ppm, Pxx_ppm); axes[2].grid(True); axes[2].set_title("PPM â€” Welch PSD"); axes[2].set_ylabel("PSD")
axes[3].semilogy(f_pcm, Pxx_pcm); axes[3].grid(True); axes[3].set_title("PCM (NRZ) â€” Welch PSD")
axes[3].set_xlabel("Frecuencia digital (ciclos/muestra)")
fig.tight_layout()
plt.show()
```

```python
# ==========================================
# CÃ“DIGO B â€” PSD con RED MLP (Keras)
# ==========================================
# Idea: generar dataset (seÃ±al -> PSD-Welch) y entrenar MLP para predecir la PSD.
# Cambia MOD âˆˆ {'PAM','PWM','PPM','PCM'} para elegir la modulaciÃ³n.

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
import tensorflow as tf
from tensorflow.keras import layers, models

# ---------- Config ----------
np.random.seed(0)
tf.random.set_seed(0)

MOD = 'PAM'         # 'PAM' | 'PWM' | 'PPM' | 'PCM'
Nsym = 96           # sÃ­mbolos por ejemplo
M    = 12           # muestras por sÃ­mbolo
Lp   = 6            # longitud pulso (<= M)
fs   = 1.0
Nfft = 512          # puntos para Welch
nperseg = 128
noverlap = 64
Ntrain = 2000       # tamaÃ±o dataset entrenamiento
Nval   = 400        # validaciÃ³n
Ntest  = 200        # test
TOTAL  = Ntrain + Nval + Ntest

# ---------- Generadores de seÃ±ales ----------
def rect_pulse(L): return np.ones(L, dtype=float)

def gen_message(Nsym):
    # mensaje suave y aleatorio (mezcla de senos + ruido pequeÃ±o)
    m = np.arange(Nsym)
    a1, a2 = np.random.uniform(0.3,0.8), np.random.uniform(0.1,0.5)
    f1, f2 = np.random.uniform(0.03,0.12), np.random.uniform(0.10,0.22)
    x = a1*np.sin(2*np.pi*f1*m) + a2*np.sin(2*np.pi*f2*m)
    x += 0.05*np.random.randn(Nsym)
    x = np.clip(x, -1.0, 1.0)
    return x

def build_signal(x_msg, MOD, M, Lp):
    p = rect_pulse(Lp)
    if MOD == 'PAM':
        s = np.zeros(Nsym*M + Lp, dtype=float)
        for i, xm in enumerate(x_msg):
            start = i*M
            s[start:start+Lp] += xm * p
        return s

    elif MOD == 'PWM':
        L0 = max(2, Lp)
        k_w = int(0.25*M)
        s = np.zeros(Nsym*M, dtype=float)
        for i, xm in enumerate(x_msg):
            Lm = int(np.round(L0 + k_w * xm))
            Lm = max(1, min(M, Lm))
            start = i*M
            s[start:start+Lm] += 1.0
        return s

    elif MOD == 'PPM':
        k_p = int(0.25*Lp)
        s = np.zeros(Nsym*M + Lp, dtype=float)
        for i, xm in enumerate(x_msg):
            shift = int(np.round(k_p * xm))
            base = i*M + (M - Lp)//2 + shift
            base = max(i*M, min(i*M + M - Lp, base))
            s[base:base+Lp] += p
        return s

    elif MOD == 'PCM':
        Qbits = np.random.choice([3,4,5])       # variar cuantizaciÃ³n
        Qlev  = 2**Qbits
        xmin, xmax = -1.0, 1.0
        xq = np.clip(x_msg, xmin, xmax)
        xq = np.round((xq - xmin)/(xmax - xmin)*(Qlev - 1)).astype(int)
        b_sym = (xq/(Qlev - 1))*2 - 1
        # a veces, quitar DC para robustecer el modelo
        if np.random.rand() < 0.5:
            b_sym = b_sym - np.mean(b_sym)
        p_b = rect_pulse(M)
        s = np.zeros(Nsym*M + M, dtype=float)
        for i, bm in enumerate(b_sym):
            start = i*M
            s[start:start+M] += bm * p_b
        return s
    else:
        raise ValueError("MOD no vÃ¡lido")

def welch_psd(x):
    f, Pxx = welch(x, fs=fs, nperseg=nperseg, noverlap=noverlap, nfft=Nfft,
                   return_onesided=True, scaling='density')
    return Pxx.astype(np.float32)

# ---------- Dataset ----------
X_list, Y_list = [], []
for _ in range(TOTAL):
    x_msg = gen_message(Nsym)
    s = build_signal(x_msg, MOD, M, Lp)
    Pxx = welch_psd(s)
    # normalizar entrada y salida para entrenamiento estable
    s_std = np.std(s) + 1e-8
    s_nrm = (s - np.mean(s)) / s_std
    Pxx_log = np.log10(Pxx + 1e-12)             # predicciÃ³n en dB (log) estabiliza
    X_list.append(s_nrm.astype(np.float32))
    Y_list.append(Pxx_log)

# padding/corte para entradas con longitudes distintas por MOD
Lmax = max(len(x) for x in X_list)
for i in range(len(X_list)):
    xi = X_list[i]
    pad = Lmax - len(xi)
    if pad > 0:
        X_list[i] = np.pad(xi, (0,pad), mode='constant')
    else:
        X_list[i] = xi[:Lmax]

X = np.stack(X_list, axis=0)           # [N, Lmax]
Y = np.stack(Y_list, axis=0)           # [N, Nfft/2+1]
print("X shape:", X.shape, "Y shape:", Y.shape)

# split
X_train, Y_train = X[:Ntrain], Y[:Ntrain]
X_val,   Y_val   = X[Ntrain:Ntrain+Nval], Y[Ntrain:Ntrain+Nval]
X_test,  Y_test  = X[Ntrain+Nval:], Y[Ntrain+Nval:]

# ---------- Modelo MLP ----------
L_in  = X.shape[1]
L_out = Y.shape[1]

model = models.Sequential([
    layers.Input(shape=(L_in,)),
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(L_out, activation='linear')
])

model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss='mse',
              metrics=[tf.keras.metrics.MeanAbsoluteError()])

hist = model.fit(X_train, Y_train,
                 validation_data=(X_val, Y_val),
                 epochs=25, batch_size=64, verbose=1)

# ---------- EvaluaciÃ³n ----------
loss, mae = model.evaluate(X_test, Y_test, verbose=0)
print(f"Test MSE: {loss:.4f}  |  Test MAE: {mae:.4f} (en dB)")

# ---------- Visualizar: pred vs target ----------
idx = np.random.randint(0, len(X_test))
y_true = Y_test[idx]
y_pred = model.predict(X_test[idx:idx+1], verbose=0)[0]

f = np.linspace(0, fs/2, L_out)
plt.figure(figsize=(10,5))
plt.plot(f, y_true, label='Welch (target)')   # dB
plt.plot(f, y_pred, label='MLP (pred)')
plt.title(f'MOD={MOD} â€” PSD(dB): Welch vs MLP')
plt.xlabel('Frecuencia (ciclos/muestra)')
plt.ylabel('PSD (dB)')
plt.grid(True)
plt.legend()
plt.show()
```
