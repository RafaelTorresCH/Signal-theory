```python
import numpy as np
import matplotlib.pyplot as plt

# ParÃ¡metros
bits = [1, 0, 1, 1, 0, 0, 1, 0]  # 8 sÃ­mbolos (d_i)
samples_per_symbol = 100  # Muestras por sÃ­mbolo
amplitude_0 = 0  # Amplitud para bit 0
amplitude_1 = 1  # Amplitud para bit 1

# Generar la seÃ±al modulada
signal = []
time = []

for i, d_i in enumerate(bits):
    # Amplitud segÃºn el bit
    a_i = amplitude_1 if d_i == 1 else amplitude_0

    # Generar pulso rectangular
    for j in range(samples_per_symbol):
        signal.append(a_i)
        time.append(i * samples_per_symbol + j)

# Convertir a arrays de numpy
signal = np.array(signal)
time = np.array(time)

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(time, signal, linewidth=2)
plt.xlabel('Tiempo (muestras)', fontsize=12)
plt.ylabel('Amplitud', fontsize=12)
plt.title('SeÃ±al Modulada: 8 SÃ­mbolos (Pulso Rectangular Ã Amplitud)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.ylim(-0.2, 1.2)

# AÃ±adir lÃ­neas verticales para separar sÃ­mbolos
for i in range(len(bits) + 1):
    plt.axvline(x=i * samples_per_symbol, color='r', linestyle='--', alpha=0.5)

# AÃ±adir etiquetas de bits
for i, bit in enumerate(bits):
    plt.text((i + 0.5) * samples_per_symbol, 1.1, f'd_{i}={bit}',
             ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

# Imprimir informaciÃ³n
print("Secuencia de bits (d_i):", bits)
print(f"NÃºmero de sÃ­mbolos: {len(bits)}")
print(f"Muestras por sÃ­mbolo: {samples_per_symbol}")
print(f"Amplitud para 0: {amplitude_0}")
print(f"Amplitud para 1: {amplitude_1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ParÃ¡metros
bits = [1, 0, 1, 1, 0, 0, 1, 0]  # 8 sÃ­mbolos (d_i)
samples_per_symbol = 100  # Muestras por sÃ­mbolo
Tb = 1  # DuraciÃ³n de cada sÃ­mbolo
fs = samples_per_symbol / Tb  # Frecuencia de muestreo
fc = 5 / Tb  # Frecuencia de portadora (5 ciclos por sÃ­mbolo)
w0 = 2 * np.pi * fc  # Frecuencia angular

# Generar la seÃ±al modulada completa
signal_complete = []
time_axis = []

for i, d_i in enumerate(bits):
    a_i = 2 * d_i - 1  # a_i = 2*d_i - 1
    t_symbol = np.linspace(i * Tb, (i + 1) * Tb, samples_per_symbol, endpoint=False)
    s_t = a_i * np.cos(w0 * t_symbol)
    signal_complete.extend(s_t)
    time_axis.extend(t_symbol)

signal_complete = np.array(signal_complete)
time_axis = np.array(time_axis)

# Generar un sÃ­mbolo individual (primer sÃ­mbolo con d_0 = 1)
d_0 = bits[0]
a_0 = 2 * d_0 - 1
t_single = np.linspace(0, Tb, samples_per_symbol, endpoint=False)
signal_single = a_0 * np.cos(w0 * t_single)

# Calcular FFT para un sÃ­mbolo
fft_single = np.fft.fft(signal_single)
fft_single_mag = np.abs(fft_single)
freq_single = np.fft.fftfreq(len(signal_single), 1/fs)

# Calcular FFT para todos los sÃ­mbolos
fft_complete = np.fft.fft(signal_complete)
fft_complete_mag = np.abs(fft_complete)
freq_complete = np.fft.fftfreq(len(signal_complete), 1/fs)

# Crear figura con subplots
fig = plt.figure(figsize=(16, 10))

# Subplot 1: SeÃ±al de un sÃ­mbolo en el tiempo
ax1 = plt.subplot(3, 2, 1)
ax1.plot(t_single, signal_single, linewidth=2, color='blue')
ax1.set_xlabel('Tiempo (s)', fontsize=10)
ax1.set_ylabel('Amplitud', fontsize=10)
ax1.set_title(f'Un SÃ­mbolo: dâ={d_0}, aâ={a_0:+d}', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)

# Subplot 2: FFT de un sÃ­mbolo (magnitud)
ax2 = plt.subplot(3, 2, 2)
ax2.plot(freq_single[:len(freq_single)//2], fft_single_mag[:len(fft_single_mag)//2],
         linewidth=2, color='red')
ax2.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax2.set_ylabel('Magnitud', fontsize=10)
ax2.set_title('FFT - Un SÃ­mbolo', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.axvline(x=fc, color='green', linestyle='--', label=f'fc={fc:.1f} Hz', linewidth=2)
ax2.legend()

# Subplot 3: SeÃ±al completa (8 sÃ­mbolos) en el tiempo
ax3 = plt.subplot(3, 2, 3)
ax3.plot(time_axis, signal_complete, linewidth=1.5, color='blue')
ax3.set_xlabel('Tiempo (s)', fontsize=10)
ax3.set_ylabel('Amplitud', fontsize=10)
ax3.set_title('8 SÃ­mbolos Completos', fontsize=12, fontweight='bold')
ax3.grid(True, alpha=0.3)
for i in range(len(bits) + 1):
    ax3.axvline(x=i * Tb, color='r', linestyle='--', alpha=0.3, linewidth=1)

# Subplot 4: FFT de todos los sÃ­mbolos (magnitud)
ax4 = plt.subplot(3, 2, 4)
ax4.plot(freq_complete[:len(freq_complete)//2], fft_complete_mag[:len(fft_complete_mag)//2],
         linewidth=1.5, color='red')
ax4.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax4.set_ylabel('Magnitud', fontsize=10)
ax4.set_title('FFT - 8 SÃ­mbolos Completos', fontsize=12, fontweight='bold')
ax4.grid(True, alpha=0.3)
ax4.axvline(x=fc, color='green', linestyle='--', label=f'fc={fc:.1f} Hz', linewidth=2)
ax4.legend()

# Subplot 5: FFT de un sÃ­mbolo (escala logarÃ­tmica)
ax5 = plt.subplot(3, 2, 5)
ax5.semilogy(freq_single[:len(freq_single)//2], fft_single_mag[:len(fft_single_mag)//2],
             linewidth=2, color='purple')
ax5.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax5.set_ylabel('Magnitud (log)', fontsize=10)
ax5.set_title('FFT - Un SÃ­mbolo (Escala Log)', fontsize=12, fontweight='bold')
ax5.grid(True, alpha=0.3, which='both')
ax5.axvline(x=fc, color='green', linestyle='--', linewidth=2)

# Subplot 6: FFT de todos los sÃ­mbolos (escala logarÃ­tmica)
ax6 = plt.subplot(3, 2, 6)
ax6.semilogy(freq_complete[:len(freq_complete)//2], fft_complete_mag[:len(fft_complete_mag)//2],
             linewidth=1.5, color='purple')
ax6.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax6.set_ylabel('Magnitud (log)', fontsize=10)
ax6.set_title('FFT - 8 SÃ­mbolos (Escala Log)', fontsize=12, fontweight='bold')
ax6.grid(True, alpha=0.3, which='both')
ax6.axvline(x=fc, color='green', linestyle='--', linewidth=2)

plt.tight_layout()
plt.show()

# Imprimir informaciÃ³n
print("="*60)
print("INFORMACIÃN DE LA SEÃAL")
print("="*60)
print(f"Secuencia de bits: {bits}")
print(f"Frecuencia de portadora fc: {fc:.2f} Hz")
print(f"Frecuencia de muestreo fs: {fs:.2f} Hz")
print(f"DuraciÃ³n por sÃ­mbolo Tb: {Tb} s")
print(f"\nUN SÃMBOLO:")
print(f"  - NÃºmero de muestras: {len(signal_single)}")
print(f"  - ResoluciÃ³n frecuencial: {fs/len(signal_single):.3f} Hz")
print(f"  - MÃ¡xima magnitud FFT: {np.max(fft_single_mag):.2f}")
print(f"\nTODOS LOS SÃMBOLOS (8):")
print(f"  - NÃºmero de muestras: {len(signal_complete)}")
print(f"  - ResoluciÃ³n frecuencial: {fs/len(signal_complete):.3f} Hz")
print(f"  - MÃ¡xima magnitud FFT: {np.max(fft_complete_mag):.2f}")
```

```python
# Reutilizando variables ya definidas anteriormente
# bits, samples_per_symbol, Tb, fs, fc, w0, signal_complete, signal_single, etc.

# Calcular Ts (perÃ­odo de sÃ­mbolo) y ancho de banda
Ts = Tb  # PerÃ­odo de sÃ­mbolo (ya definido como Tb = 1)
Rs = 1 / Ts  # Tasa de sÃ­mbolos (sÃ­mbolos por segundo)

# Ancho de banda de cada sÃ­mbolo
# Para seÃ±al BPSK: BW â 2*Rs = 2/Ts
# El ancho de banda del lÃ³bulo principal es 2/Ts
BW_main_lobe = 2 / Ts  # Ancho de banda del lÃ³bulo principal
BW_null_to_null = 2 * Rs  # Ancho de banda de nulo a nulo

# Calcular densidad espectral de potencia (PSD) para mejor visualizaciÃ³n
# Para un sÃ­mbolo
fft_single_normalized = fft_single_mag / len(signal_single)
psd_single = (fft_single_normalized ** 2)

# Para todos los sÃ­mbolos
fft_complete_normalized = fft_complete_mag / len(signal_complete)
psd_complete = (fft_complete_normalized ** 2)

# Crear figura mejorada con informaciÃ³n de Ts y BW
fig = plt.figure(figsize=(16, 12))

# Subplot 1: SeÃ±al de un sÃ­mbolo
ax1 = plt.subplot(4, 2, 1)
ax1.plot(t_single, signal_single, linewidth=2, color='blue')
ax1.set_xlabel('Tiempo (s)', fontsize=10)
ax1.set_ylabel('Amplitud', fontsize=10)
ax1.set_title(f'Un SÃ­mbolo: Ts = {Ts} s, Rs = {Rs} sÃ­mbolos/s', fontsize=11, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axvline(x=0, color='red', linestyle='--', alpha=0.5)
ax1.axvline(x=Ts, color='red', linestyle='--', alpha=0.5, label=f'Ts={Ts}s')
ax1.legend()

# Subplot 2: FFT de un sÃ­mbolo con marcas de BW
ax2 = plt.subplot(4, 2, 2)
ax2.plot(freq_single[:len(freq_single)//2], fft_single_mag[:len(fft_single_mag)//2],
         linewidth=2, color='red')
ax2.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax2.set_ylabel('Magnitud', fontsize=10)
ax2.set_title(f'FFT Un SÃ­mbolo - BW = {BW_main_lobe:.2f} Hz', fontsize=11, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.axvline(x=fc, color='green', linestyle='-', label=f'fc={fc:.1f} Hz', linewidth=2)
ax2.axvline(x=fc - Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5)
ax2.axvline(x=fc + Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5, label=f'BWâ{BW_main_lobe:.1f} Hz')
ax2.legend(fontsize=8)
ax2.set_xlim(0, 15)

# Subplot 3: SeÃ±al completa
ax3 = plt.subplot(4, 2, 3)
ax3.plot(time_axis, signal_complete, linewidth=1.5, color='blue')
ax3.set_xlabel('Tiempo (s)', fontsize=10)
ax3.set_ylabel('Amplitud', fontsize=10)
ax3.set_title(f'8 SÃ­mbolos: DuraciÃ³n Total = {8*Ts} s', fontsize=11, fontweight='bold')
ax3.grid(True, alpha=0.3)
for i in range(len(bits) + 1):
    ax3.axvline(x=i * Ts, color='r', linestyle='--', alpha=0.3, linewidth=1)
    if i < len(bits):
        ax3.text((i + 0.5) * Ts, 1.2, f'Ts', ha='center', fontsize=8)

# Subplot 4: FFT de todos los sÃ­mbolos con marcas de BW
ax4 = plt.subplot(4, 2, 4)
ax4.plot(freq_complete[:len(freq_complete)//2], fft_complete_mag[:len(fft_complete_mag)//2],
         linewidth=1.5, color='red')
ax4.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax4.set_ylabel('Magnitud', fontsize=10)
ax4.set_title(f'FFT 8 SÃ­mbolos - BW = {BW_main_lobe:.2f} Hz', fontsize=11, fontweight='bold')
ax4.grid(True, alpha=0.3)
ax4.axvline(x=fc, color='green', linestyle='-', label=f'fc={fc:.1f} Hz', linewidth=2)
ax4.axvline(x=fc - Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5)
ax4.axvline(x=fc + Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5, label=f'BWâ{BW_main_lobe:.1f} Hz')
ax4.legend(fontsize=8)
ax4.set_xlim(0, 15)

# Subplot 5: PSD de un sÃ­mbolo (escala log)
ax5 = plt.subplot(4, 2, 5)
ax5.semilogy(freq_single[:len(freq_single)//2], psd_single[:len(psd_single)//2],
             linewidth=2, color='purple')
ax5.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax5.set_ylabel('PSD (log)', fontsize=10)
ax5.set_title('Densidad Espectral - Un SÃ­mbolo', fontsize=11, fontweight='bold')
ax5.grid(True, alpha=0.3, which='both')
ax5.axvline(x=fc, color='green', linestyle='-', linewidth=2)
ax5.axvline(x=fc - Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5)
ax5.axvline(x=fc + Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5)
ax5.set_xlim(0, 15)

# Subplot 6: PSD de todos los sÃ­mbolos (escala log)
ax6 = plt.subplot(4, 2, 6)
ax6.semilogy(freq_complete[:len(freq_complete)//2], psd_complete[:len(psd_complete)//2],
             linewidth=1.5, color='purple')
ax6.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax6.set_ylabel('PSD (log)', fontsize=10)
ax6.set_title('Densidad Espectral - 8 SÃ­mbolos', fontsize=11, fontweight='bold')
ax6.grid(True, alpha=0.3, which='both')
ax6.axvline(x=fc, color='green', linestyle='-', linewidth=2)
ax6.axvline(x=fc - Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5)
ax6.axvline(x=fc + Rs, color='orange', linestyle='--', alpha=0.7, linewidth=1.5)
ax6.set_xlim(0, 15)

# Subplot 7: Espectro alrededor de fc (zoom) - Un sÃ­mbolo
ax7 = plt.subplot(4, 2, 7)
ax7.plot(freq_single[:len(freq_single)//2], fft_single_mag[:len(fft_single_mag)//2],
         linewidth=2, color='red')
ax7.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax7.set_ylabel('Magnitud', fontsize=10)
ax7.set_title(f'Zoom en fc Â± BW/2 - Un SÃ­mbolo', fontsize=11, fontweight='bold')
ax7.grid(True, alpha=0.3)
ax7.axvline(x=fc, color='green', linestyle='-', linewidth=2)
ax7.axvspan(fc - Rs, fc + Rs, alpha=0.2, color='yellow', label=f'BW={BW_main_lobe:.1f} Hz')
ax7.set_xlim(fc - 3, fc + 3)
ax7.legend(fontsize=8)

# Subplot 8: Espectro alrededor de fc (zoom) - Todos los sÃ­mbolos
ax8 = plt.subplot(4, 2, 8)
ax8.plot(freq_complete[:len(freq_complete)//2], fft_complete_mag[:len(fft_complete_mag)//2],
         linewidth=1.5, color='red')
ax8.set_xlabel('Frecuencia (Hz)', fontsize=10)
ax8.set_ylabel('Magnitud', fontsize=10)
ax8.set_title(f'Zoom en fc Â± BW/2 - 8 SÃ­mbolos', fontsize=11, fontweight='bold')
ax8.grid(True, alpha=0.3)
ax8.axvline(x=fc, color='green', linestyle='-', linewidth=2)
ax8.axvspan(fc - Rs, fc + Rs, alpha=0.2, color='yellow', label=f'BW={BW_main_lobe:.1f} Hz')
ax8.set_xlim(fc - 3, fc + 3)
ax8.legend(fontsize=8)

plt.tight_layout()
plt.show()

# Tabla de resultados
print("="*70)
print("PARÃMETROS DE SÃMBOLO Y ANCHO DE BANDA")
print("="*70)
print(f"\nð TEMPORALES:")
print(f"  Ts (PerÃ­odo de sÃ­mbolo)          : {Ts} s")
print(f"  Rs (Tasa de sÃ­mbolos)            : {Rs} sÃ­mbolos/s")
print(f"  DuraciÃ³n total (8 sÃ­mbolos)      : {8*Ts} s")
print(f"  fc (Frecuencia portadora)        : {fc} Hz")
print(f"  Ïâ (Frecuencia angular)          : {w0:.2f} rad/s")

print(f"\nð¡ ESPECTRALES:")
print(f"  BW (LÃ³bulo principal)            : {BW_main_lobe} Hz")
print(f"  BW = 2/Ts = 2ÃRs                 : {BW_null_to_null} Hz")
print(f"  Rango de frecuencias principales : [{fc-Rs:.1f}, {fc+Rs:.1f}] Hz")
print(f"  BW ocupado (fc Â± Rs)             : {2*Rs} Hz")

print(f"\nð RESOLUCIÃN FFT:")
print(f"  Un sÃ­mbolo ({len(signal_single)} muestras)  : Îf = {fs/len(signal_single):.3f} Hz")
print(f"  8 sÃ­mbolos ({len(signal_complete)} muestras) : Îf = {fs/len(signal_complete):.3f} Hz")

print(f"\nð¡ OBSERVACIONES:")
print(f"  - El ancho de banda del sÃ­mbolo BPSK es BW = 2/Ts = {BW_main_lobe} Hz")
print(f"  - El 90% de la potencia estÃ¡ concentrada en fc Â± Rs = [{fc-Rs}, {fc+Rs}] Hz")
print(f"  - La seÃ±al de 8 sÃ­mbolos tiene mejor resoluciÃ³n frecuencial")
print(f"  - Los lÃ³bulos laterales decaen como 1/f (visible en escala log)")
print("="*70)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Recrear la seÃ±al (basado en la celda 1 de tu cuaderno) ---
bits = [1, 0, 1, 1, 0, 0, 1, 0]  # 8 sÃ­mbolos
samples_per_symbol = 100
amplitude_0 = 0
amplitude_1 = 1

signal = []
for d_i in bits:
    a_i = amplitude_1 if d_i == 1 else amplitude_0
    signal.extend([a_i] * samples_per_symbol)

signal = np.array(signal)
N = len(signal) # NÃºmero total de muestras

# --- 2. Calcular parÃ¡metros de frecuencia (basado en la celda 17) ---
Ts_ms = 125.0
Ts_s = Ts_ms / 1000.0  # 0.125 s
Rs = 1.0 / Ts_s         # 8.0 Hz (Tasa de sÃ­mbolos)

# Frecuencia de muestreo (Fs) = Muestras por sÃ­mbolo * Tasa de sÃ­mbolos
Fs = samples_per_symbol * Rs  # 100 * 8.0 = 800.0 Hz

# --- 3. Calcular la FFT de la seÃ±al ---
# Normalizamos por N para obtener las amplitudes correctas
fft_mag = np.abs(np.fft.fft(signal)) / N
freq_axis = np.fft.fftfreq(N, d=1/Fs)

# Centrar la FFT (para que f=0 estÃ© en el medio)
fft_mag_shifted = np.fft.fftshift(fft_mag)
freq_axis_shifted = np.fft.fftshift(freq_axis)

# --- 4. Encontrar los picos ---

# a) Pico del LÃ³bulo Principal (en f=0)
# El promedio de la seÃ±al [1,0,1,1,0,0,1,0] es (4 * 1 + 4 * 0) / 8 = 0.5
dc_index = np.argmin(np.abs(freq_axis_shifted - 0.0))
main_lobe_peak_mag = fft_mag_shifted[dc_index]

# b) Pico del LÃ³bulo Lateral
# Buscamos el mÃ¡ximo en las frecuencias |f| > Rs (es decir, |f| > 8.0 Hz)
side_lobe_indices = np.where(np.abs(freq_axis_shifted) > Rs)
side_lobe_peak_mag = np.max(fft_mag_shifted[side_lobe_indices])

# Encontrar la frecuencia exacta de ese pico lateral
side_lobe_peak_freq = freq_axis_shifted[side_lobe_indices][np.argmax(fft_mag_shifted[side_lobe_indices])]

# --- 5. Mostrar Resultados ---
print(f"ParÃ¡metros de anÃ¡lisis:")
print(f"  Rs = {Rs} Hz (Ancho del lÃ³bulo principal: +/- {Rs} Hz)")
print(f"  Fs = {Fs} Hz")
print("-" * 40)
print(f"Pico del LÃ³bulo Principal (f=0 Hz): {main_lobe_peak_mag:.4f} (Amplitud)")
print(f"Pico del LÃ³bulo Lateral (f={side_lobe_peak_freq:.2f} Hz): {side_lobe_peak_mag:.4f} (Amplitud)")
print("-" * 40)

# Calcular la atenuaciÃ³n en decibelios (dB)
attenuation_db = 20 * np.log10(side_lobe_peak_mag / main_lobe_peak_mag)
print(f"AtenuaciÃ³n del LÃ³bulo Lateral: {attenuation_db:.2f} dB")
print(f"(Valor teÃ³rico esperado: ~ -13.26 dB)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

def raised_cosine_pulse(t, Ts, alpha):
    """
    Genera el pulso de Coseno Alzado (no-causal) en el dominio del tiempo.
    Maneja las singularidades en t=0 y t = +/- Ts/(2*alpha).
    """

    # Evita divisiÃ³n por cero si alpha es 0 (caso Sinc)
    if alpha == 0:
        return np.sinc(t / Ts)

    # Normalizar tiempo
    t_norm = t / Ts

    # --- Manejar casos especiales (singularidades) ---

    # 1. Caso t = 0 (Pico principal)
    # El lÃ­mite (usando L'HÃ´pital) es 1.0
    val_t0 = 1.0

    # 2. Caso t = +/- Ts / (2*alpha)
    # El lÃ­mite es (pi/4) * sinc(1 / (2*alpha))
    t_sing = Ts / (2.0 * alpha)
    val_sing = (np.pi / 4.0) * np.sinc(1.0 / (2.0 * alpha))

    # --- FÃ³rmula general ---
    sinc_part = np.sinc(t_norm)
    cos_part = np.cos(np.pi * alpha * t_norm)
    den_part = 1.0 - (2.0 * alpha * t_norm)**2

    # Usar np.where para aplicar la fÃ³rmula o los valores lÃ­mite

    # Primero, la fÃ³rmula general (con un pequeÃ±o Ã©psilon para evitar 0/0)
    pulse = sinc_part * (cos_part / (den_part + 1e-20))

    # Sobrescribir los puntos de singularidad
    # (usamos np.isclose para precisiÃ³n de punto flotante)

    # Sobrescribir t=0
    pulse[np.isclose(t, 0.0)] = val_t0

    # Sobrescribir t = +/- Ts / (2*alpha)
    pulse[np.isclose(np.abs(t), t_sing)] = val_sing

    return pulse

def analyze_lobes(t, pulse, label):
    """
    Encuentra el pico principal, el primer lÃ³bulo lateral y calcula
    la atenuaciÃ³n en dB.
    """
    # Encontrar picos en la magnitud
    # (min_height evita picos de ruido numÃ©rico muy pequeÃ±os)
    peaks_idx, _ = find_peaks(np.abs(pulse), height=1e-5)
    peak_values = np.abs(pulse[peaks_idx])

    if len(peak_values) < 2:
        print(f"--- {label} ---")
        print("No se encontraron suficientes picos para el anÃ¡lisis.")
        return

    # Ordenar picos de mayor a menor
    sort_idx = np.argsort(peak_values)[::-1]

    main_peak_val = peak_values[sort_idx[0]]
    side_lobe_peak_val = peak_values[sort_idx[1]]

    # Encontrar el Ã­ndice original de los picos
    main_peak_t_idx = peaks_idx[sort_idx[0]]
    side_peak_t_idx = peaks_idx[sort_idx[1]]

    # Calcular en dB
    sll_db = 20 * np.log10(side_lobe_peak_val / main_peak_val)

    print(f"--- {label} ---")
    print(f"Pico LÃ³bulo Principal: {main_peak_val:.4f} (en t={t[main_peak_t_idx]:.2f})")
    print(f"Pico 1er LÃ³bulo Lateral: {side_lobe_peak_val:.4f} (en t={t[side_peak_t_idx]:.2f})")
    print(f"Nivel del LÃ³bulo Lateral (SLL): {sll_db:.2f} dB")

# --- ParÃ¡metros de simulaciÃ³n ---
Ts = 1.0  # Tiempo de sÃ­mbolo (normalizado a 1)
# Rango de tiempo: +/- 6 sÃ­mbolos para ver bien los lÃ³bulos
t = np.linspace(-6 * Ts, 6 * Ts, 4001) # MÃ¡s puntos para mejor precisiÃ³n

# --- GrÃ¡ficos y AnÃ¡lisis ---
plt.figure(figsize=(12, 8))

print("AnÃ¡lisis de LÃ³bulos Laterales (Dominio del Tiempo):\n")

# Caso 1: alpha = 0 (Pulso Sinc)
try:
    alpha_0 = 0.0
    pulse_0 = raised_cosine_pulse(t, Ts, alpha_0)
    analyze_lobes(t, pulse_0, "Alpha = 0 (Sinc)")
    # Graficamos en dB
    plt.plot(t, 20*np.log10(np.abs(pulse_0)), label=f'Alpha = {alpha_0} (Sinc)', linestyle='--')
except Exception as e:
    print(f"Error analizando Alpha=0: {e}")

# Caso 2: alpha = 0.5
try:
    alpha_1 = 0.5
    pulse_1 = raised_cosine_pulse(t, Ts, alpha_1)
    analyze_lobes(t, pulse_1, "\nAlpha = 0.5")
    plt.plot(t, 20*np.log10(np.abs(pulse_1)), label=f'Alpha = {alpha_1}')
except Exception as e:
    print(f"Error analizando Alpha=0.5: {e}")

# Caso 3: alpha = 1.0
try:
    alpha_2 = 1.0
    pulse_2 = raised_cosine_pulse(t, Ts, alpha_2)
    analyze_lobes(t, pulse_2, "\nAlpha = 1.0")
    plt.plot(t, 20*np.log10(np.abs(pulse_2)), label=f'Alpha = {alpha_2}')
except Exception as e:
    print(f"Error analizando Alpha=1.0: {e}")

# ConfiguraciÃ³n del grÃ¡fico (en dB)
plt.title('Pulso de Coseno Alzado (Dominio del Tiempo) - Escala dB')
plt.xlabel('Tiempo (normalizado por Ts)')
plt.ylabel('Magnitud (dB)')
plt.ylim(-60, 5) # LÃ­mite inferior para ver los lÃ³bulos
plt.grid(True, which='both', linestyle=':')
plt.legend()
plt.tight_layout() # Ajustar etiquetas
plt.savefig('raised_cosine_pulses_db.png')

print(f"\nGrÃ¡fico 'raised_cosine_pulses_db.png' generado.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- ParÃ¡metros de tu Cuaderno ---
Ts_ms = 125.0
Ts_s = Ts_ms / 1000.0 # 0.125 s
Rs = 1.0 / Ts_s # 8.0 sÃ­mbolos/s
samples_per_symbol = 100
dt = Ts_s / samples_per_symbol

# --- 1. SÃ­mbolos de entrada (d_i) ---
symbols_d_i = [0, 2, 3, 0, 1, 1, 2]

# --- 2. AsignaciÃ³n (Mapping) a a_i = I + jQ ---
mapping_a_i = {
    0: -1 + 0j,  # d_i=0 -> I=-1, Q=0
    1: 0 - 1j,  # d_i=1 -> I=0,  Q=-1
    2: 1 + 0j,   # d_i=2 -> I=1,  Q=0
    3: 0 + 1j   # d_i=3 -> I=0,  Q=1
}

# --- 3. Generar la SeÃ±al Modulada Compleja (Banda Base) ---
# Esta seÃ±al (signal_modulated) es la s_BB(t) = I(t) + jQ(t)
signal_modulated = []
time = []
current_time = 0.0

print(f"--- Generando SeÃ±al RF (QPSK) ---")
print(f"Secuencia (d_i): {symbols_d_i}")

for d_i in symbols_d_i:
    a_i = mapping_a_i.get(d_i) # a_i es complejo (I + jQ)
    for _ in range(samples_per_symbol):
        signal_modulated.append(a_i)
        time.append(current_time)
        current_time += dt

signal_modulated = np.array(signal_modulated, dtype=np.complex128)
time = np.array(time)

# Extraemos las seÃ±ales I(t) y Q(t) (que son pulsos rectangulares)
I_t = signal_modulated.real
Q_t = signal_modulated.imag

print("SeÃ±ales I(t) y Q(t) (banda base) generadas.")

# --- 4. Definir la Portadora de RF (fc) ---
# Usamos la fc = Rs/2 = 4.0 Hz definida en la Celda 17 de tu cuaderno
fc = Rs / 2.0
wc = 2 * np.pi * fc # Frecuencia angular

print(f"Tasa de SÃ­mbolos (Rs): {Rs} Hz")
print(f"Frecuencia Portadora (fc = Rs/2): {fc} Hz")

# --- 5. Modular a RF ---
# FÃ³rmula de modulaciÃ³n QPSK:
# s_RF(t) = I(t) * cos(wc*t) - Q(t) * sin(wc*t)

# Generar las portadoras (Coseno y Seno)
carrier_cos = np.cos(wc * time)
carrier_sin = np.sin(wc * time)

# Multiplicar I(t) por el Coseno
I_component_RF = I_t * carrier_cos
# Multiplicar Q(t) por el Seno
Q_component_RF = Q_t * carrier_sin

# Restar las componentes (I*cos - Q*sin)
signal_RF = I_component_RF - Q_component_RF

print("SeÃ±al RF generada (s_RF = I*cos - Q*sin).")

# --- 6. GrÃ¡fica de la SeÃ±al RF ---
plt.figure(figsize=(12, 10))

# --- Plot 1: Componente I(t) * Coseno ---
plt.subplot(3, 1, 1)
plt.plot(time, I_component_RF, label='Componente I (I*cos)', color='blue')
# Graficar la envolvente I(t) (el pulso rectangular)
plt.plot(time, I_t, color='blue', linestyle='--', alpha=0.7, label='Envolvente I(t)')
plt.title(f'SeÃ±al Modulada RF (fc = {fc} Hz)')
plt.ylabel('Amplitud (I)')
plt.grid(True, linestyle=':')
plt.legend(loc='upper right')

# --- Plot 2: Componente -Q(t) * Seno ---
plt.subplot(3, 1, 2)
# Graficamos -Q*sin, que es lo que se suma
plt.plot(time, -Q_component_RF, label='Componente Q (-Q*sin)', color='red')
# Graficar la envolvente -Q(t)
plt.plot(time, -Q_t, color='red', linestyle='--', alpha=0.7, label='Envolvente -Q(t)')
plt.ylabel('Amplitud (Q)')
plt.grid(True, linestyle=':')
plt.legend(loc='upper right')

# --- Plot 3: SeÃ±al RF Final (Suma de las dos) ---
plt.subplot(3, 1, 3)
plt.plot(time, signal_RF, label='SeÃ±al RF Final (I*cos - Q*sin)', color='purple')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud RF')
plt.grid(True, linestyle=':')
plt.legend(loc='upper right')

plt.tight_layout()
plt.savefig('qpsk_rf_signal.png')
print("GrÃ¡fica 'qpsk_rf_signal.png' generada.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- ParÃ¡metros de tu Cuaderno ---
Ts_ms = 125.0
Ts_s = Ts_ms / 1000.0 # 0.125 s
Rs = 1.0 / Ts_s # 8.0 sÃ­mbolos/s
samples_per_symbol = 100
dt = Ts_s / samples_per_symbol

# --- 1. SÃ­mbolos de entrada (d_i) ---
symbols_d_i = [0, 2, 3, 0, 1, 1, 2]

# --- 2. AsignaciÃ³n (Mapping) a a_i = I + jQ ---
mapping_a_i = {
    0: -1 + 0j,  # d_i=0 -> I=-1, Q=0
    1: 0 - 1j,  # d_i=1 -> I=0,  Q=-1
    2: 1 + 0j,   # d_i=2 -> I=1,  Q=0
    3: 0 + 1j   # d_i=3 -> I=0,  Q=1
}

# --- 3. Generar la SeÃ±al Modulada Compleja (Banda Base) ---
# s_BB(t) = I(t) + jQ(t)
signal_modulated = []
time = []
current_time = 0.0

print(f"--- Generando SeÃ±al de TransmisiÃ³n")

for d_i in symbols_d_i:
    a_i = mapping_a_i.get(d_i)
    for _ in range(samples_per_symbol):
        signal_modulated.append(a_i)
        time.append(current_time)
        current_time += dt

signal_modulated = np.array(signal_modulated, dtype=np.complex128)
time = np.array(time)

# I(t) y Q(t) son las seÃ±ales (pulsos rectangulares) que modulan
I_t = signal_modulated.real
Q_t = signal_modulated.imag

print("SeÃ±ales I(t) y Q(t) (banda base) generadas.")

# --- 4. Definir la Portadora de RF (fc) ---
fc = Rs / 2.0 # 4.0 Hz
wc = 2 * np.pi * fc
print(f"Frecuencia Portadora (fc): {fc} Hz")

# --- 5. Modular a RF ---
# Esta es la fÃ³rmula de transmisiÃ³n QPSK
# s_RF(t) = I(t) * cos(wc*t) - Q(t) * sin(wc*t)
carrier_cos = np.cos(wc * time)
carrier_sin = np.sin(wc * time)

# ESTA ES LA SEÃAL QUE VERÃA EL OSCILOSCOPIO:
signal_RF = I_t * carrier_cos - Q_t * carrier_sin

# La envolvente A(t) = sqrt(I(t)^2 + Q(t)^2)
envelope = np.sqrt(I_t**2 + Q_t**2)

print("SeÃ±al RF (TransmisiÃ³n) calculada.")

# --- 6. GrÃ¡fica (Lo que verÃ­a el Osciloscopio) ---
plt.figure(figsize=(12, 6))

# Graficar la SeÃ±al RF (la modulada)
plt.plot(time, signal_RF, label='SeÃ±al RF (Vista de Osciloscopio)', color='purple', linewidth=1.5)

# Graficar la Envolvente (para referencia)
plt.plot(time, envelope, label='Envolvente (sqrt(I^2+Q^2))', color='black', linestyle='--', linewidth=2)
plt.plot(time, -envelope, color='black', linestyle='--', linewidth=2)

plt.title(f'SeÃ±al de TransmisiÃ³n RF (fc = {fc} Hz)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud (V, E-Field, etc.)')
plt.grid(True, linestyle=':')
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('qpsk_rf_oscilloscope_view.png')
print("GrÃ¡fica 'qpsk_rf_oscilloscope_view.png' generada.")
```
